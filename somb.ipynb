```python
fig,ax=plt.subplots()
fig.subplots_adjust(bottom=0.25)

axsld = fig.add_axes((0.2, 0.05,0.65, 0.075))    
sl=Slider(
    ax=axsld,
    label='a',
    valmin=0,
    valmax=4,
    valstep=0.1,
    valinit=1
)

x=np.arange(-2,2,0.01)
y=x**4-0*x**2
ln,=ax.plot(x,y,'b')

ax.set_title('$y=x^4-ax^2$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')


sl.on_changed(update)
plt.show ()
```
